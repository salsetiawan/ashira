```python
# import Library yang dibutuhkan
import pandas as pd                   #dataframe manipulation
import numpy as np                    #scientific camputing
import seaborn as sns                 #visualization
import matplotlib.pyplot as plt       #visualization
%matplotlib inline
sns.set(color_codes=True)

import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Files

nb = pd.read_csv("C:/Users/HP/Desktop/Data IMDb/name.basics.csv") 
tb = pd.read_csv("C:/Users/HP/Desktop/Data IMDb/title.basics.csv") 
tc = pd.read_csv("C:/Users/HP/Desktop/Data IMDb/title.crew.csv") 
te = pd.read_csv("C:/Users/HP/Desktop/Data IMDb/title.episode.csv") 
tr = pd.read_csv("C:/Users/HP/Desktop/Data IMDb/title.ratings.csv") 
```

```python
''' Join semua tabel '''

# Join tabel name.basics dan title.crew
nb.head()
```

```python
tc.head()
```

```python
# Merge pertama: untuk directors
dw = tc.merge(nb, left_on='directors', right_on='nameID', how='left')\
          .rename(columns={'name': 'directorName'})

# Merge kedua: untuk writers
dw = dw.merge(nb, left_on='writers', right_on='nameID', how='left')\
           .rename(columns={'name': 'writerName'})

# Drop kolom yang tidak dibutuhkan
dw = dw[['movieID', 'directorName', 'writerName']]

dw
```

```python
# Join tabel title.basics, title.ratings, title.episode, dw (director.writer)
tb.head()
```

```python
tr.head()
```

```python
te.head()
```

```python
# Merge pertama dw ke te
sd = te.merge(dw, left_on='tconst', right_on='movieID', how='left')

# Merge kedua tb
sd = sd.merge(tb, left_on='tconst', right_on='movieID', how='left')

# Merge kedua tr
sd = sd.merge(tr, left_on='tconst', right_on='tconst', how='left')

# Drop kolom yang tidak dibutuhkan
sd = sd[['tconst', 'parentTconst', 'seasonNumber', 'episodeNumber',
        'directorName', 'writerName',
         'type',	'movieName',	'year',	'duration', 'genres',
        'averageRating',	'numVotes'
        ]]

sd
```

```python
# Cek Missing Values
print(sd.isnull().sum())
```

```python
''' movieNamenya NaN dihapus '''

sd = sd.dropna(subset=['movieName'])
sd
```

```python
''' Buat data frame baru lagi Movie.detail (md)'''

# Tables yang akan digunakan
'''
director.writer (dw)
title.basics.tsv.gz
title.ratings.tsv.gz
'''
# Merge pertama dw ke tb
md = tb.merge(dw, left_on='movieID', right_on='movieID', how='left')

# Merge kedua tr
md = md.merge(tr, left_on='movieID', right_on='tconst', how='left')

# Drop kolom yang tidak dibutuhkan
md = md[['movieID', 'directorName', 'writerName',
         'type',	'movieName',	'year',	'duration', 'genres',
        'averageRating',	'numVotes'
        ]]

md
```
